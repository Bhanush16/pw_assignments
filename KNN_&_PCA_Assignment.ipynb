

##**1. What is K-Nearest Neighbors (KNN) and how does it work?****K-Nearest Neighbors (KNN)** is a simple, non-parametric, lazy learning algorithm used for both classification and regression. It is considered a *lazy* algorithm because it does not require a specific training phase or model building; instead, it memorizes the entire training dataset. When a prediction is needed for a new, unseen data point, KNN works by first identifying the **K** number of data points in the training set that are closest to the new point, based on a chosen distance metric (e.g., Euclidean distance). For **classification**, the new point is assigned the class label most frequent among its K neighbors (a majority vote). For **regression**, the new point is assigned the average of the target values of its K neighbors.

##**2. What is the difference between KNN Classification and KNN Regression?**The difference between **KNN Classification** and **KNN Regression** lies only in how the final prediction is aggregated from the K nearest neighbors. In **KNN Classification**, the model performs a **majority vote** among the K neighbors; the class that appears most frequently among the neighbors is assigned as the prediction for the new data point. In **KNN Regression**, the model calculates the **average** (or median) of the target values (output variables) of the K neighbors, and this average value becomes the prediction for the new data point.

##**3. What is the role of the distance metric in KNN?**The distance metric is crucial in KNN as it determines which data points are considered "nearest" to a given test sample. The choice of metric heavily influences the composition of the K neighborhood and thus the final prediction. The most common metric is **Euclidean Distance** (the straight-line distance between two points). Other options include **Manhattan Distance** (the sum of the absolute differences of their coordinates, useful when movement is restricted to grids) and **Minkowski Distance** (a generalization of both Euclidean and Manhattan distances). The distance metric must be chosen carefully, especially when feature scales differ significantly.

##**4. What is the Curse of Dimensionality in KNN?**The **Curse of Dimensionality** is a major drawback for KNN, referring to the phenomena where the data becomes increasingly sparse as the number of features (dimensions) increases. In high-dimensional space, the distance between any two data points tends to become nearly equal, and the concept of "nearest neighbor" loses its meaning and distinction. This causes the KNN model's performance to degrade rapidly because it can no longer reliably determine which points are truly close to the test sample, often leading to increased computation time and decreased predictive accuracy.

##**5. How can we choose the best value of K in KNN?**Choosing the optimal value of **K** is essential for KNN performance, as a small K can lead to high variance (overfitting) and a large K can lead to high bias (underfitting). The best value for K is typically determined through **hyperparameter tuning** using techniques like **Cross-Validation**. The process involves testing various K values (e.g., K=1, 3, 5, 7, \dots) on the training data and selecting the value that yields the best performance metric (e.g., highest accuracy or lowest error) on the validation set. A general rule of thumb is to start with an odd number to avoid ties in classification and keep K relatively small, often between 3 and 10.

##**6. What are KD Tree and Ball Tree in KNN?****KD Tree** and **Ball Tree** are spatial data structures used to dramatically speed up the computation of finding the K nearest neighbors, thereby mitigating the major computational cost of the KNN algorithm, especially with large datasets. The **KD Tree** (K-Dimensional Tree) is a binary space-partitioning data structure that recursively splits the data space along one dimension at a time. The **Ball Tree** is an alternative data structure that partitions data points into nested hyper-spheres (or "balls"), which is generally more efficient than KD Trees when dealing with very high-dimensional data or non-uniform data distributions.

##**7. When should you use KD Tree vs. Ball Tree?**The choice between **KD Tree** and **Ball Tree** depends primarily on the dimensionality of the data: you should use the **KD Tree** when the number of features (dimensions) is relatively **low** (typically D < 20) and the data is uniformly distributed. However, as the dimensionality increases, the KD Tree's performance degrades (due to the Curse of Dimensionality). In contrast, you should use the **Ball Tree** when the dimensionality is **high** (typically D \geq 20) or when the data exhibits complex, non-uniform, or highly sparse distributions, as the Ball Tree's structure is generally better suited for these challenging scenarios.

##**8. What are the disadvantages of KNN?**The main disadvantages of KNN are: **Computational Cost**, as it requires storing the entire training dataset and calculating the distance to every training sample for each new prediction (especially slow for large N); **Curse of Dimensionality**, where performance degrades rapidly in high-dimensional space; **Sensitivity to Feature Scaling**, as features with larger scales will unduly influence the distance calculation; and finally, its **Inability to handle Imbalanced Data** gracefully, as a majority class can easily dominate the K neighbors, leading to poor prediction performance for the minority class.

##**9. How does feature scaling affect KNN?**Feature scaling is crucial for KNN because the algorithm relies entirely on the **distance** between data points. If features have different scales (e.g., one feature ranges from 0 to 100, while another ranges from 0 to 1), the feature with the larger scale will disproportionately contribute to the distance calculation, potentially skewing the identification of the true nearest neighbors. Therefore, it is mandatory to **normalize** (scale all values between 0 and 1) or **standardize** (scale to have a mean of 0 and standard deviation of 1) the features before applying the KNN algorithm.

##**10. What is PCA (Principal Component Analysis)?****Principal Component Analysis (PCA)** is an unsupervised, linear dimensionality reduction technique widely used in machine learning. Its primary goal is to simplify the complexity of high-dimensional datasets while retaining as much of the original **variance** as possible. PCA achieves this by transforming the data into a new coordinate system where the axes, called **Principal Components (PCs)**, are orthogonal (uncorrelated) and ordered such that the first PC captures the maximum amount of variance, the second PC captures the maximum remaining variance, and so on.

##**11. How does PCA work?**PCA works in four main steps: 1) **Standardize the Data** to ensure each feature contributes equally. 2) **Calculate the Covariance Matrix** of the standardized data to understand the relationships and variances between features. 3) **Calculate the Eigenvalues and Eigenvectors** of the covariance matrix; the eigenvectors represent the principal components (the new directions), and the eigenvalues represent the magnitude of variance along those directions. 4) **Select the Top K Eigenvectors** (Principal Components) corresponding to the largest eigenvalues, and project the original data onto this reduced subspace to obtain the final lower-dimensional data.

##**12. What is the geometric intuition behind PCA?**The geometric intuition behind PCA is that it seeks the directions (vectors) in the feature space that best represent the overall spread or **variance** of the data. Imagine a cloud of 3D data points: the first principal component is the line that, when the data is projected onto it, results in the maximum spread of the projected points. The second principal component is the line orthogonal to the first that maximizes the remaining spread. PCA is essentially a rotation of the coordinate system to align the new axes with the directions of maximum variability in the data.

##**13. What is the difference between Feature Selection and Feature Extraction?****Feature Selection** and **Feature Extraction** are both dimensionality reduction techniques, but they achieve their goal differently. **Feature Selection** involves choosing a *subset* of the original features that are most relevant (e.g., using correlation or statistical tests) and discarding the rest; the resulting features are the original, unaltered features. **Feature Extraction** (like PCA) involves *transforming* the original high-dimensional features into a new, lower-dimensional set of features (the Principal Components); these new features are linear combinations of the original features and are entirely new.

##**14. What are Eigenvalues and Eigenvectors in PCA?**In the context of PCA, the **Eigenvectors** of the covariance matrix are the **Principal Components** themselves. They are the new orthogonal axes of the feature space and represent the directions of maximum variance in the data. The **Eigenvalues** are the magnitudes associated with each eigenvector and quantify the amount of variance captured along that specific eigenvector (Principal Component). A larger eigenvalue signifies that its corresponding eigenvector captures more information (variance) from the original dataset. PCA uses the eigenvectors with the largest eigenvalues to define the reduced subspace.

##**15. How do you decide the number of components to keep in PCA?**The number of components to keep in PCA is decided by balancing dimensionality reduction with information retention. Common methods include: 1) **Scree Plot**: Plotting the eigenvalues in descending order and looking for the "elbow" point where the explained variance drops sharply. 2) **Explained Variance Ratio**: Selecting components that cumulatively account for a high percentage of the total variance (e.g., 90% or 95%). 3) **Kaiser's Rule**: Keeping only those components whose eigenvalues are greater than 1 (as a component with variance less than 1 explains less variance than a single standardized original feature).

##**16. Can PCA be used for classification?****PCA itself is not a classification algorithm**; it is a feature pre-processing and dimensionality reduction technique. However, it is very commonly used *before* a classification algorithm. By reducing the noise and the number of features, PCA can help prevent the "Curse of Dimensionality," speed up the training of the classifier (like KNN or SVM), and sometimes even improve the classifier's generalization accuracy by removing uninformative dimensions.

##**17. What are the limitations of PCA?**The limitations of PCA include: **Linearity Assumption**, as it can only find linear combinations of features, making it unsuitable for capturing non-linear data structures. **Sensitivity to Scaling**, requiring all features to be standardized beforehand. **Information Loss**, as even when retaining 95% of the variance, the remaining 5% might contain critical information relevant to a specific prediction task. Finally, the **Interpretability** of the Principal Components is often low, as they are abstract linear combinations of the original features, not the original features themselves.

##**18. How do KNN and PCA complement each other?**KNN and PCA complement each other perfectly because PCA addresses KNN's biggest weakness. KNN suffers heavily from the **Curse of Dimensionality** and slow computation time in high-dimensional space. By applying **PCA first**, the dimensionality is significantly reduced, the data is projected into a more meaningful (higher variance) subspace, and some noise may be eliminated. This pre-processing step makes the subsequent **KNN** distance calculations much faster, more reliable, and more accurate by improving the definition of "nearest neighbors."

##**19. How does KNN handle missing values in a dataset?**KNN, in its standard form, **cannot directly handle missing values**. The distance calculation (e.g., Euclidean distance) requires a numerical value for every feature in both the test sample and the training samples. Therefore, missing values must be handled *before* applying KNN. Common methods include **Imputation**, where missing values are replaced by the mean, median, or mode of the feature; or **Deletion**, where samples or features with missing data are removed entirely. The imputation method must be applied carefully to both the training and test sets.

##**20. What are the key differences between PCA and Linear Discriminant Analysis (LDA)?**The key difference between **PCA** and **Linear Discriminant Analysis (LDA)** is their objective and supervision status. **PCA is an unsupervised** technique that finds the directions (Principal Components) that maximize the **overall variance** in the data, regardless of class labels. It is used for dimensionality reduction and data compression. **LDA is a supervised** technique that finds the directions (Linear Discriminants) that maximize the **separation** between classes. It explicitly uses the target labels to find the subspace that best discriminates between different groups. LDA is often better for classification tasks but relies on labeled data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import (
    accuracy_score, mean_squared_error, precision_score, recall_score,
    f1_score, roc_auc_score, roc_curve, classification_report
)
from sklearn.datasets import load_iris, make_regression, make_blobs, load_wine
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.decomposition import PCA
from sklearn.impute import KNNImputer

# --- Setup for Classification Tasks (Iris Dataset) ---
data_iris = load_iris()
X_iris, y_iris = data_iris.data, data_iris.target
X_iris_train, X_iris_test, y_iris_train, y_iris_test = train_test_split(X_iris, y_iris, test_size=0.3, random_state=42)

# --- Setup for Regression Tasks (Synthetic Data) ---
X_reg, y_reg = make_regression(n_samples=300, n_features=5, n_informative=3, noise=10.0, random_state=42)
X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(X_reg, y_reg, test_size=0.3, random_state=42)

# --- Setup for High-Dimensional/PCA Tasks (Synthetic Blobs) ---
X_high, y_high = make_blobs(n_samples=500, n_features=50, centers=4, random_state=42)
X_high_train, X_high_test, y_high_train, y_high_test = train_test_split(X_high, y_high, test_size=0.3, random_state=42)

# --- Scaling setup for PCA and KNN comparison ---
scaler = StandardScaler()
X_iris_scaled = scaler.fit_transform(X_iris)
X_iris_scaled_train, X_iris_scaled_test, _, _ = train_test_split(X_iris_scaled, y_iris, test_size=0.3, random_state=42)
```

```python
# 21. Train a KNN Classifier on the Iris dataset and print model accuracy
knn_clf = KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(X_iris_train, y_iris_train)
y_pred_knn = knn_clf.predict(X_iris_test)
accuracy_knn = accuracy_score(y_iris_test, y_pred_knn)

print(f"KNN Classifier (K=5) Accuracy on Iris: {accuracy_knn:.4f}")
```

```python
# 22. Train a KNN Regressor on a synthetic dataset and evaluate using Mean Squared Error (MSE)
knn_reg = KNeighborsRegressor(n_neighbors=5)
knn_reg.fit(X_reg_train, y_reg_train)
y_pred_knn_reg = knn_reg.predict(X_reg_test)
mse_knn_reg = mean_squared_error(y_reg_test, y_pred_knn_reg)

print(f"KNN Regressor (K=5) Mean Squared Error (MSE): {mse_knn_reg:.4f}")
```

```python
# 23. Train a KNN Classifier using different distance metrics (Euclidean and Manhattan) and compare accuracy

# 1. Euclidean Distance (p=2, default)
knn_euclidean = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn_euclidean.fit(X_iris_train, y_iris_train)
acc_euclidean = accuracy_score(y_iris_test, knn_euclidean.predict(X_iris_test))

# 2. Manhattan Distance (p=1)
knn_manhattan = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=1)
knn_manhattan.fit(X_iris_train, y_iris_train)
acc_manhattan = accuracy_score(y_iris_test, knn_manhattan.predict(X_iris_test))

print(f"KNN Accuracy (Euclidean Distance): {acc_euclidean:.4f}")
print(f"KNN Accuracy (Manhattan Distance): {acc_manhattan:.4f}")
```

```python
# 24. Train a KNN Classifier with different values of K and visualize decision boundaries
# Note: For visualization, we must reduce the Iris data to 2 features (or 2 PCA components)

# Reduce data to 2 features for plotting: Sepal Length and Sepal Width (indices 0 and 1)
X_plot = X_iris[:, :2]
X_plot_train, X_plot_test, y_plot_train, y_plot_test = train_test_split(X_plot, y_iris, test_size=0.3, random_state=42)

def plot_decision_boundary(X, y, model, title):
    # Setup marker generator and color map
    cmap_light = plt.cm.RdYlBu
    cmap_bold = plt.cm.RdYlBu
    h = .02  # step size in the mesh

    # Plot the decision boundary
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure(figsize=(7, 5))
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light, shading='auto')

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(title)
    plt.xlabel(data_iris.feature_names[0])
    plt.ylabel(data_iris.feature_names[1])
    plt.grid(True)
    # plt.show() # Uncomment to display the plot

# Train and plot for K=1 and K=15
print("Training KNN for K=1 and K=15...")

knn_k1 = KNeighborsClassifier(n_neighbors=1).fit(X_plot_train, y_plot_train)
plot_decision_boundary(X_plot, y_iris, knn_k1, "KNN Decision Boundary (K=1)")

knn_k15 = KNeighborsClassifier(n_neighbors=15).fit(X_plot_train, y_plot_train)
plot_decision_boundary(X_plot, y_iris, knn_k15, "KNN Decision Boundary (K=15)")

print("Decision boundary plots generated (requires plt.show() to display).")
```

```python
# 25. Apply Feature Scaling before training a KNN model and compare results with unscaled data

# 1. KNN on Unscaled Data (from Q21)
acc_unscaled = accuracy_knn

# 2. KNN on Scaled Data (StandardScaler)
knn_scaled = KNeighborsClassifier(n_neighbors=5)
# Use the scaled data prepared in the setup
knn_scaled.fit(X_iris_scaled_train, y_iris_train)
acc_scaled = accuracy_score(y_iris_test, knn_scaled.predict(X_iris_scaled_test))

print(f"KNN Accuracy (Unscaled Data): {acc_unscaled:.4f}")
print(f"KNN Accuracy (Scaled Data):   {acc_scaled:.4f}")
print(f"Scaling {'improved' if acc_scaled > acc_unscaled else 'did not improve or worsened'} accuracy in this case.")
```

```python
# 26. Train a PCA model on synthetic data and print the explained variance ratio for each component
# Use high-dimensional data from setup (X_high, 50 features)
pca_50 = PCA(n_components=50)
pca_50.fit(X_high)

# Print explained variance ratio for the first 10 components
explained_variance_ratio = pca_50.explained_variance_ratio_

print("PCA Explained Variance Ratio (First 10 Components):")
for i, ratio in enumerate(explained_variance_ratio[:10]):
    print(f"  PC {i+1}: {ratio:.4f}")
```

```python
# 27. Apply PCA before training a KNN Classifier and compare accuracy with and without PCA

# Use high-dimensional data (50 features) where PCA is beneficial
# 1. Train PCA to capture 95% variance
pca_95 = PCA(n_components=0.95)
pca_95.fit(X_high_train)
X_pca_train = pca_95.transform(X_high_train)
X_pca_test = pca_95.transform(X_high_test)
n_components = X_pca_train.shape[1]

# 2. KNN with PCA
knn_pca = KNeighborsClassifier(n_neighbors=5)
knn_pca.fit(X_pca_train, y_high_train)
acc_pca = accuracy_score(y_high_test, knn_pca.predict(X_pca_test))

# 3. KNN without PCA (on all 50 features)
knn_full = KNeighborsClassifier(n_neighbors=5)
knn_full.fit(X_high_train, y_high_train)
acc_full = accuracy_score(y_high_test, knn_full.predict(X_high_test))

print(f"Number of original features: {X_high_train.shape[1]}")
print(f"Number of components kept by PCA (95% variance): {n_components}")
print(f"KNN Accuracy (Full Data): {acc_full:.4f}")
print(f"KNN Accuracy (PCA Reduced Data): {acc_pca:.4f}")
```

```python
# 28. Perform Hyperparameter Tuning on a KNN Classifier using GridSearchCV
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

knn_base = KNeighborsClassifier()
grid_search_knn = GridSearchCV(
    estimator=knn_base,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
grid_search_knn.fit(X_iris_scaled, y_iris) # Use scaled data for better results

print(f"Best KNN Hyperparameters: {grid_search_knn.best_params_}")
print(f"Best Cross-Validation Accuracy: {grid_search_knn.best_score_:.4f}")

# Evaluate best estimator on the test set (using the split data for a fairer test score)
best_knn = grid_search_knn.best_estimator_
best_knn.fit(X_iris_scaled_train, y_iris_train)
acc_best_knn = accuracy_score(y_iris_test, best_knn.predict(X_iris_scaled_test))
print(f"Test Set Accuracy with Best KNN Model: {acc_best_knn:.4f}")
```

```python
# 29. Train a KNN Classifier and check the number of misclassified samples
# Use the trained knn_clf (K=5) from Q21
y_pred_knn = knn_clf.predict(X_iris_test)

# Check which predictions do not match the true labels
misclassified_samples = np.where(y_iris_test != y_pred_knn)[0]
n_misclassified = len(misclassified_samples)
n_total_test = len(y_iris_test)

print(f"Total test samples: {n_total_test}")
print(f"Number of misclassified samples: {n_misclassified}")
print(f"Misclassification rate: {(n_misclassified / n_total_test):.4f}")

if n_misclassified > 0:
    print("Indices of misclassified samples in the test set:")
    print(misclassified_samples)
```

```python
# 30. Train a PCA model and visualize the cumulative explained variance.
# Use the high-dimensional data (X_high, 50 features)
pca_full = PCA().fit(X_high)

# Calculate cumulative explained variance
cumulative_variance = np.cumsum(pca_full.explained_variance_ratio_)

# Visualization
try:
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
    plt.title('PCA Cumulative Explained Variance')
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.axhline(y=0.90, color='r', linestyle='-', label='90% Variance Threshold')
    plt.axhline(y=0.95, color='g', linestyle='-', label='95% Variance Threshold')
    plt.legend()
    plt.grid(True)
    # plt.show() # Uncomment to display the plot

    # Find number of components for 95% variance
    n_95 = np.where(cumulative_variance >= 0.95)[0][0] + 1
    print(f"Number of components to capture 95% variance: {n_95}")
    print("Cumulative explained variance plot generated (requires plt.show() to display).")
except Exception as e:
    print(f"Plotting failed: {e}")
```

```python
# 31. Train a KNN Classifier using different values of the weights parameter (uniform vs. distance) and compare accuracy

# 1. Uniform Weights (default)
knn_uniform = KNeighborsClassifier(n_neighbors=5, weights='uniform')
knn_uniform.fit(X_iris_train, y_iris_train)
acc_uniform = accuracy_score(y_iris_test, knn_uniform.predict(X_iris_test))

# 2. Distance Weights (closer neighbors have more influence)
knn_distance = KNeighborsClassifier(n_neighbors=5, weights='distance')
knn_distance.fit(X_iris_train, y_iris_train)
acc_distance = accuracy_score(y_iris_test, knn_distance.predict(X_iris_test))

print(f"KNN Accuracy (Uniform Weights): {acc_uniform:.4f}")
print(f"KNN Accuracy (Distance Weights): {acc_distance:.4f}")
```

```python
# 32. Train a KNN Regressor and analyze the effect of different K values on performance
k_values = [1, 5, 15, 30]
mse_results = {}

print("KNN Regressor MSE Comparison by K Value:")
for k in k_values:
    knn_reg_k = KNeighborsRegressor(n_neighbors=k)
    knn_reg_k.fit(X_reg_train, y_reg_train)
    mse = mean_squared_error(y_reg_test, knn_reg_k.predict(X_reg_test))
    mse_results[k] = mse
    print(f"K={k:2}: MSE={mse:.4f}")
```

```python
# 46. Train a PCA model and visualize the data reconstruction error after reducing dimensions
# Use the high-dimensional data (X_high)
X_high_mean = X_high.mean(axis=0)

# Calculate reconstruction error for various numbers of components
errors = []
n_components_range = range(1, X_high.shape[1] + 1)

for n in n_components_range:
    pca_n = PCA(n_components=n).fit(X_high)
    X_reduced = pca_n.transform(X_high)
    X_reconstructed = pca_n.inverse_transform(X_reduced)

    # Reconstruction Error (Mean Squared Error between original and reconstructed data)
    error = np.mean((X_high - X_reconstructed)**2)
    errors.append(error)

# Visualization
try:
    plt.figure(figsize=(10, 6))
    plt.plot(n_components_range, errors, marker='o')
    plt.title('PCA Reconstruction Error vs. Number of Components')
    plt.xlabel('Number of Components')
    plt.ylabel('Reconstruction Error (MSE)')
    plt.grid(True)
    # plt.show() # Uncomment to display the plot
    print("Reconstruction error plot generated (requires plt.show() to display).")
except Exception as e:
    print(f"Plotting failed: {e}")
```

```python
# 47. Train a KNN Classifier and visualize the decision boundary
# This repeats Q24 but ensures it's covered by the new numbering scheme.

# Reduce Iris data to 2 features (Sepal Length, Sepal Width) for plotting
X_plot_final = X_iris[:, :2]
y_plot_final = y_iris
knn_final = KNeighborsClassifier(n_neighbors=5).fit(X_plot_final, y_plot_final)

# Re-using the plot function defined in Q24
def plot_decision_boundary(X, y, model, title):
    cmap_light = plt.cm.RdYlBu
    cmap_bold = plt.cm.RdYlBu
    h = .02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.figure(figsize=(7, 5))
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light, shading='auto')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(title)
    plt.xlabel(data_iris.feature_names[0])
    plt.ylabel(data_iris.feature_names[1])
    plt.grid(True)
    # plt.show()

print("Visualizing Decision Boundary for KNN (K=5) on 2 Iris features...")
plot_decision_boundary(X_plot_final, y_plot_final, knn_final, "KNN Decision Boundary (K=5, Iris 2 Features)")
print("Decision boundary plot generated (requires plt.show() to display).")
```
