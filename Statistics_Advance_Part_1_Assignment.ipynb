#Statistics Advance Part 1


**1. What is a random variable in probability theory?**
**A)** A random variable is a variable whose value is determined by the outcome of a random experiment. It assigns numerical values to each possible outcome of a probabilistic process.

---

**2. What are the types of random variables?**
**A)** There are two main types:

* **Discrete Random Variables**: Take on countable values (e.g., number of heads in coin tosses).
* **Continuous Random Variables**: Take on any value within a range or interval (e.g., height, temperature).

---

**3. What is the difference between discrete and continuous distributions?**
**A)**

* **Discrete Distributions**: Deal with countable outcomes (e.g., Binomial, Poisson).
* **Continuous Distributions**: Deal with outcomes that take on infinite values in an interval (e.g., Normal, Uniform).

---

**4. What are probability distribution functions (PDF)?**
**A)** A **Probability Distribution Function** defines the probability of each possible value a random variable can take. For continuous variables, PDF describes the density of probability over an interval.

---

**5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**
**A)**

* **PDF** gives the probability at a specific value (for discrete) or density (for continuous).
* **CDF** gives the cumulative probability that the variable is less than or equal to a certain value.

---

**6. What is a discrete uniform distribution?**
**A)** A discrete uniform distribution has equal probability for all outcomes in a finite set. Example: rolling a fair six-sided die (P(1) = P(2) = ... = P(6) = 1/6).

---

**7. What are the key properties of a Bernoulli distribution?**
**A)**

* Only two outcomes: success (1) or failure (0)
* Defined by one parameter: $p$, the probability of success
* Mean: $p$, Variance: $p(1-p)$

---

**8. What is the binomial distribution, and how is it used in probability?**
**A)** The binomial distribution models the number of successes in $n$ independent trials, each with probability $p$ of success.
Formula:

$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

---

**9. What is the Poisson distribution and where is it applied?**
**A)** The Poisson distribution models the number of events occurring in a fixed interval of time or space with a known average rate and independent occurrences.
Used in: call centers, traffic flow, etc.

---

**10. What is a continuous uniform distribution?**
**A)** A distribution where all values within a given interval are equally likely.
PDF:

$$
f(x) = \frac{1}{b - a},\ \text{for } a \le x \le b
$$

---

**11. What are the characteristics of a normal distribution?**
**A)**

* Bell-shaped, symmetric around the mean
* Mean = Median = Mode
* Fully defined by mean (μ) and standard deviation (σ)
* Total area under curve = 1

---

**12. What is the standard normal distribution, and why is it important?**
**A)** The standard normal distribution is a normal distribution with a mean of 0 and standard deviation of 1. It is used for Z-score calculations and standardizing data.

---

**13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
**A)** CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's distribution.
It allows for inference and hypothesis testing using normal theory.

---

**14. How does the Central Limit Theorem relate to the normal distribution?**
**A)** CLT justifies using the **normal distribution** to approximate the behavior of sample means from any population when sample size is large (typically $n \ge 30$).

---

**15. What is the application of Z statistics in hypothesis testing?**
**A)** Z-statistics are used to test hypotheses about population means when the population standard deviation is known. It helps determine whether a sample mean significantly differs from a known population mean.

---

**16. How do you calculate a Z-score, and what does it represent?**
**A)**

$$
Z = \frac{X - \mu}{\sigma}
$$

It represents how many standard deviations a data point $X$ is from the mean $\mu$. A Z-score helps identify how unusual a value is.

---

**17. What are point estimates and interval estimates in statistics?**
**A)**

* **Point Estimate**: A single value (e.g., sample mean) that serves as an estimate of a population parameter.
* **Interval Estimate**: A range (confidence interval) that likely contains the population parameter.

---

**18. What is the significance of confidence intervals in statistical analysis?**
**A)** Confidence intervals provide a range of values with an associated confidence level (e.g., 95%) that likely contains the true population parameter, offering more information than a single point estimate.

---

**19. What is the relationship between a Z-score and a confidence interval?**
**A)** Z-scores determine the margin of error for confidence intervals when the population standard deviation is known.
For example:

$$
CI = \bar{x} \pm Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}
$$

---

**20. How are Z-scores used to compare different distributions?**
**A)** Z-scores standardize values from different distributions, allowing comparison on the same scale (mean = 0, std dev = 1), regardless of original units.

---

**21. What are the assumptions for applying the Central Limit Theorem?**
**A)**

* Random sampling
* Independent observations
* Finite variance
* Sample size should be sufficiently large (usually $n \ge 30$)

---

**22. What is the concept of expected value in a probability distribution?**
**A)** The expected value (or mean) of a random variable is the long-run average outcome of a random process.

$$
E[X] = \sum x_i P(x_i)\ \text{(for discrete)}\quad \text{or}\quad \int x f(x) dx\ \text{(for continuous)}
$$

---

**23. How does a probability distribution relate to the expected outcome of a random variable?**
**A)** A probability distribution defines the likelihood of all possible outcomes. The **expected value** is computed using this distribution and represents the theoretical average over many trials.

```python
#Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not 0 <= p <= 1:
        raise ValueError("Probability p must be between 0 and 1")

    return p if x == 1 else 1 - p

p = 0.7
print("P(X=1):", bernoulli_pdf(1, p))
print("P(X=0):", bernoulli_pdf(0, p))
```

```python
#Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std

    plt.figure(figsize=(8, 5))
    plt.plot(z_scores, marker='o', linestyle='', color='purple')
    plt.axhline(y=0, color='black', linestyle='--')
    plt.axhline(y=2, color='red', linestyle='--', label='Z = ±2')
    plt.axhline(y=-2, color='red', linestyle='--')
    plt.title("Z-Scores of Dataset")
    plt.xlabel("Index")
    plt.ylabel("Z-Score")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    return z_scores
data = np.random.normal(100, 15, 100)
z = calculate_and_plot_zscores(data)
```

```python
#Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():

    x = np.linspace(-4, 4, 1000)

    pdf = norm.pdf(x, loc=0, scale=1)


    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, color='blue', label='Standard Normal PDF')
    plt.fill_between(x, pdf, alpha=0.3, color='skyblue')

    for z in [-2, -1, 0, 1, 2]:
        plt.axvline(x=z, linestyle='--', color='gray', alpha=0.6)
        plt.text(z, norm.pdf(z), f'z={z}', fontsize=8, ha='center', va='bottom')

    plt.title("Standard Normal Distribution (μ = 0, σ = 1)")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()
plot_standard_normal_distribution()
```

```python
#Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    return (x - mean) / std_dev
data_point = 85
mean = 70
std_dev = 10
z_score = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for data point {data_point}: {z_score:.2f}")
x_vals = np.linspace(-4, 4, 1000)
pdf_vals = norm.pdf(x_vals)
plt.figure(figsize=(8, 5))
plt.plot(x_vals, pdf_vals, label='Standard Normal Distribution', color='blue')
plt.axvline(x=z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.fill_between(x_vals, 0, pdf_vals, where=(x_vals <= z_score), color='red', alpha=0.3)
plt.title("Z-Score Comparison with Standard Normal Distribution")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
#Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

# Sample data (can be randomly generated or real)
sample = np.array([72, 68, 75, 70, 69, 73, 71, 74, 76, 68,
                   72, 67, 70, 69, 71, 73, 74, 68, 72, 70,
                   69, 75, 73, 71, 72, 70, 69, 74, 73, 68,
                   72, 70, 71, 70, 69, 68, 72, 74, 73, 71,
                   69, 70, 68, 73, 71, 70, 69, 74, 75, 70])
population_mean = 70
population_std = 10
sample_mean = np.mean(sample)
sample_size = len(sample)
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Statistic: {z_statistic:.3f}")
print(f"P-Value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H₀)")
else:
    print("Fail to reject the null hypothesis (H₀)")
```

```python
#Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats

# Sample dataset
data = np.array([72, 68, 75, 70, 69, 73, 71, 74, 76, 68,
                 72, 67, 70, 69, 71, 73, 74, 68, 72, 70,
                 69, 75, 73, 71, 72, 70, 69, 74, 73, 68])
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
sample_size = len(data)
confidence = 0.95
alpha = 1 - confidence
t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1)
margin_error = t_critical * (sample_std / np.sqrt(sample_size))
ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error
print(f"Sample Mean: {sample_mean:.2f}")
print(f"{int(confidence * 100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
import scipy.stats as stats
np.random.seed(42)
mean = 100
std_dev = 15
sample_size = 50
data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
confidence_level = 0.95
alpha = 1 - confidence_level
df = sample_size - 1
t_critical = stats.t.ppf(1 - alpha / 2, df)
margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error
print(f"Sample Mean: {sample_mean:.2f}")
print(f"{int(confidence_level * 100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution_skew_kurtosis(size=1000, mean=0, std=1):
    data = np.random.normal(loc=mean, scale=std, size=size)
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)

    print(f"Skewness: {data_skewness:.4f}")
    print(f"Kurtosis: {data_kurtosis:.4f}")

    return data_skewness, data_kurtosis
analyze_distribution_skew_kurtosis()
```
