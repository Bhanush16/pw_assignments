

**1. What is hypothesis testing in statistics**\*
**A)** Hypothesis testing is a statistical method used to make decisions about population parameters based on sample data. It helps determine whether there is enough evidence to reject a hypothesis about a population.

---

**2. What is the null hypothesis, and how does it differ from the alternative hypothesis**\*
**A)** The **null hypothesis (H₀)** assumes no effect or no difference in the population. The **alternative hypothesis (H₁)** suggests there is a significant effect or difference. Hypothesis testing aims to decide whether to reject H₀ in favor of H₁.

---

**3. What is the significance level in hypothesis testing, and why is it important**\*
**A)** The **significance level (α)** is the probability of rejecting a true null hypothesis (Type I error). Common values are 0.05 or 0.01. It sets the threshold for deciding whether an observed result is statistically significant.

---

**4. What does a P-value represent in hypothesis testing**\*
**A)** The **P-value** is the probability of obtaining test results at least as extreme as the observed ones, assuming the null hypothesis is true. A smaller P-value indicates stronger evidence against the null hypothesis.

---

**5. How do you interpret the P-value in hypothesis testing**\*
**A)** If **P-value < α**, reject the null hypothesis (evidence suggests a significant result).
If **P-value ≥ α**, fail to reject the null hypothesis (insufficient evidence of a significant result).

---

**6. What are Type 1 and Type 2 errors in hypothesis testing**\*
**A)**

* **Type I error**: Rejecting a true null hypothesis (false positive).
* **Type II error**: Failing to reject a false null hypothesis (false negative).

---

**7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing**\*
**A)**

* **One-tailed test**: Tests for a deviation in one direction (greater than or less than).
* **Two-tailed test**: Tests for deviations in both directions (not equal).

---

**8. What is the Z-test, and when is it used in hypothesis testing**\*
**A)** A **Z-test** is used to determine if there is a significant difference between sample and population means when the population standard deviation is known and sample size is large (n > 30).

---

**9. How do you calculate the Z-score, and what does it represent in hypothesis testing**\*
**A)**

$$
Z = \frac{\bar{x} - \mu}{\sigma/\sqrt{n}}
$$

It shows how many standard deviations a sample mean is from the population mean.

---

**10. What is the T-distribution, and when should it be used instead of the normal distribution**\*
**A)** The **T-distribution** is used when the population standard deviation is unknown and the sample size is small (n < 30). It's wider than the normal distribution to account for variability.

---

**11. What is the difference between a Z-test and a T-test**\*
**A)**

* **Z-test**: Used when population standard deviation is known.
* **T-test**: Used when population standard deviation is unknown.

---

**12. What is the T-test, and how is it used in hypothesis testing**\*
**A)** A **T-test** compares sample means to test hypotheses about population means. Types include:

* **One-sample T-test**
* **Independent two-sample T-test**
* **Paired sample T-test**

---

**13. What is the relationship between Z-test and T-test in hypothesis testing**\*
**A)** Both test hypotheses about means. The **Z-test** assumes known standard deviation and large sample size; the **T-test** is used when standard deviation is unknown or sample size is small.

---

**14. What is a confidence interval, and how is it used to interpret statistical results**\*
**A)** A **confidence interval** gives a range of values within which a population parameter is likely to fall. A 95% CI means we are 95% confident the true mean lies in that range.

---

**15. What is the margin of error, and how does it affect the confidence interval**\*
**A)** The **margin of error** is half the width of the confidence interval. A larger margin of error means a wider interval and less precision in the estimate.

---

**16. How is Bayes' Theorem used in statistics, and what is its significance**\*
**A)** **Bayes' Theorem** updates the probability of a hypothesis based on new evidence:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

It's useful in decision-making, spam filtering, diagnostics, etc.

---

**17. What is the Chi-square distribution, and when is it used**\*
**A)** The **Chi-square distribution** is used for testing relationships between categorical variables and goodness-of-fit tests. It is right-skewed and defined only for positive values.

---

**18. What is the Chi-square goodness of fit test, and how is it applied**\*
**A)** It tests whether an observed frequency distribution differs from an expected one. If the test statistic is greater than the critical value, we reject the null hypothesis.

---

**19. What is the F-distribution, and when is it used in hypothesis testing**\*
**A)** The **F-distribution** is used to compare two variances. It is used in ANOVA and the F-test. It’s skewed right and only takes positive values.

---

**20. What is an ANOVA test, and what are its assumptions**\*
**A)** **ANOVA (Analysis of Variance)** compares means of 3+ groups to see if at least one mean is different. Assumptions: independence, normality, and equal variance (homogeneity).

---

**21. What are the different types of ANOVA tests**\*
**A)**

* **One-way ANOVA**: One factor, multiple groups.
* **Two-way ANOVA**: Two factors (can assess interaction).
* **Repeated Measures ANOVA**: Same subjects over time or conditions.

---

**22. What is the F-test, and how does it relate to hypothesis testing**\*
**A)** The **F-test** compares two variances to determine if they are significantly different. It is the basis for ANOVA and helps test model significance in regression.

```python
# Implement a one-sample Z-test using Python to compare the sample mean with the population mean
def one_sample_z_test(data, pop_mean, pop_std):
    sample_mean = np.mean(data)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(len(data)))
    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

sample_data = np.random.normal(104, 10, 30)
z, p = one_sample_z_test(sample_data, 100, 10)
print("Z-score:", z)
print("P-value:", p)
```

```python
#Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

sample_mean = 105
population_mean = 100
population_std = 15
sample_size = 50

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Z-critical (two-tailed): ±{z_critical:.2f}")

if abs(z_score) > z_critical:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")

z_values = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(z_values, 0, 1)

plt.figure(figsize=(10, 5))
plt.plot(z_values, y, label='Standard Normal Distribution', color='blue')
plt.fill_between(z_values, y, where=(z_values <= -z_critical), color='red', alpha=0.5, label='Rejection Region (left)')
plt.fill_between(z_values, y, where=(z_values >= z_critical), color='red', alpha=0.5, label='Rejection Region (right)')
plt.axvline(z_score, color='green', linestyle='--', label=f'Z = {z_score:.2f}')
plt.axvline(-z_critical, color='black', linestyle=':', label='-Z critical')
plt.axvline(z_critical, color='black', linestyle=':', label='Z critical')

plt.title('Two-Tailed Z-Test: Decision Region')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(alpha=0.05, effect_size=1, std_dev=1, sample_size=30):
    se = std_dev / np.sqrt(sample_size)
    x = np.linspace(-4, 6, 1000)

    null_dist = norm(loc=0, scale=se)
    alt_dist = norm(loc=effect_size, scale=se)

    z_critical = norm.ppf(1 - alpha)
    critical_value = z_critical * se

    type1_x = np.linspace(critical_value, 6, 500)
    type2_x = np.linspace(-4, critical_value, 500)

    plt.figure(figsize=(10, 6))
    plt.plot(x, null_dist.pdf(x), label='Null Hypothesis (H₀)', color='blue')
    plt.plot(x, alt_dist.pdf(x), label='Alternative Hypothesis (H₁)', color='green')

    plt.fill_between(type1_x, null_dist.pdf(type1_x), color='red', alpha=0.4, label='Type I Error (α)')
    plt.fill_between(type2_x, alt_dist.pdf(type2_x), color='orange', alpha=0.4, label='Type II Error (β)')

    plt.axvline(critical_value, color='black', linestyle='--', label='Critical Value')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    beta = alt_dist.cdf(critical_value)
    power = 1 - beta

    print(f"Critical value: {critical_value:.2f}")
    print(f"Type I error (alpha): {alpha}")
    print(f"Type II error (beta): {beta:.4f}")
    print(f"Power of the test (1 - beta): {power:.4f}")
```

```python
#Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)
    margin = t_critical * std_err
    lower = mean - margin
    upper = mean + margin
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print(f"Sample Mean: {mean:.2f}")
    print("Interpretation: We are", int(confidence*100), "% confident that the true population mean lies within this interval.")

# Example usage
sample_data = [52, 55, 60, 58, 57, 54, 59, 61, 56, 53]
confidence_interval(sample_data, confidence=0.95)
```

```python
#Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy.stats import t

data = [52, 55, 60, 58, 57, 54, 59, 61, 56, 53]
confidence = 0.95

n = len(data)
mean = np.mean(data)
std_err = np.std(data, ddof=1) / np.sqrt(n)
t_critical = t.ppf((1 + confidence) / 2, df=n - 1)
margin_of_error = t_critical * std_err

print(f"Sample Mean: {mean:.2f}")
print(f"Confidence Level: {int(confidence*100)}%")
print(f"Margin of Error: {margin_of_error:.2f}")
```

```python
#D Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    numerator = likelihood_B_given_A * prior_A
    denominator = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A)
    posterior = numerator / denominator
    return posterior

P_D = 0.01
P_not_D = 0.99
P_Pos_given_D = 0.99
P_Pos_given_not_D = 0.05

posterior = bayes_theorem(P_D, P_Pos_given_D, P_not_D, P_Pos_given_not_D)
print(f"Probability of having the disease given a positive test: {posterior:.4f}")
```

```python
#Perform a Chi-square test for independence between two categorical variables in Python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

data = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Preference': ['Tea', 'Tea', 'Coffee', 'Coffee', 'Tea', 'Coffee', 'Coffee', 'Tea']
})

contingency_table = pd.crosstab(data['Gender'], data['Preference'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Contingency Table:")
print(contingency_table)
print(f"\nChi-square Statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns))

alpha = 0.05
if p < alpha:
    print("\nResult: Reject the null hypothesis (variables are dependent).")
else:
    print("\nResult: Fail to reject the null hypothesis (variables are independent).")
```

```python
#Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution
import numpy as np
from scipy.stats import chisquare

observed = np.array([40, 35, 25])

expected_proportions = np.array([1/3, 1/3, 1/3])
expected = expected_proportions * np.sum(observed)

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (observed does not fit expected distribution).")
else:
    print("Result: Fail to reject the null hypothesis (observed fits expected distribution).")

```

```python
#Implement an F-test using Python to compare the variances of two random samples
import numpy as np
from scipy.stats import f

np.random.seed(0)

sample1 = np.random.normal(loc=50, scale=5, size=30)
sample2 = np.random.normal(loc=50, scale=10, size=30)

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

f_stat = var1 / var2 if var1 > var2 else var2 / var1
df1 = len(sample1) - 1
df2 = len(sample2) - 1

p_value = 2 * (1 - f.cdf(f_stat, df1, df2)) if var1 > var2 else 2 * f.cdf(f_stat, df1, df2)

print(f"Variance of Sample 1: {var1:.4f}")
print(f"Variance of Sample 2: {var2:.4f}")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (variances are significantly different).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference in variances).")
```

```python
# Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import numpy as np
from scipy.stats import shapiro, levene
import matplotlib.pyplot as plt

def check_anova_assumptions(*groups):
    print("Normality (Shapiro-Wilk Test):")
    for i, group in enumerate(groups, start=1):
        stat, p = shapiro(group)
        result = "Pass" if p > 0.05 else "Fail"
        print(f"  Group {i}: W={stat:.4f}, p={p:.4f} → {result}")

    stat, p = levene(*groups)
    result = "Pass" if p > 0.05 else "Fail"
    print("\nEqual Variance (Levene’s Test):")
    print(f"  Levene Stat={stat:.4f}, p={p:.4f} → {result}")

    print("\nIndependence: Assumed by design (e.g., random, unpaired samples)")
    plt.boxplot(groups, labels=[f'Group {i+1}' for i in range(len(groups))])
    plt.title('Boxplot for Visual Independence & Variance Check')
    plt.grid(True)
    plt.show()

group1 = [23, 20, 22, 21, 24]
group2 = [30, 28, 29, 31, 32]
group3 = [40, 42, 41, 39, 43]
check_anova_assumptions(group1, group2, group3)

```

```python
# Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.DataFrame({
    'Score': [88, 92, 85, 90, 95, 89, 91, 87, 84, 86,
              78, 80, 76, 79, 75, 77, 74, 73, 72, 70,
              68, 70, 65, 67, 66, 64, 69, 63, 62, 61],
    'TeachingMethod': ['A']*10 + ['B']*10 + ['C']*10,
    'Gender': ['Male', 'Female']*15
})
model = ols('Score ~ C(TeachingMethod) * C(Gender)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("\nTwo-Way ANOVA Results:")
print(anova_table)
plt.figure(figsize=(8, 5))
sns.boxplot(x='TeachingMethod', y='Score', hue='Gender', data=data)
plt.title('Scores by Teaching Method and Gender')
plt.grid(True)
plt.show()
```

```python
#D Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means
import numpy as np
import pandas as pd
from scipy.stats import f_oneway
import matplotlib.pyplot as plt
import seaborn as sns

group1 = [23, 20, 22, 21, 24]
group2 = [30, 28, 29, 31, 32]
group3 = [40, 42, 41, 39, 43]


data = pd.DataFrame({
    'Score': group1 + group2 + group3,
    'Group': ['Group 1']*len(group1) + ['Group 2']*len(group2) + ['Group 3']*len(group3)
})
f_stat, p_value = f_oneway(group1, group2, group3)

print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (at least one group mean is different).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference among group means).")
plt.figure(figsize=(8, 5))
sns.boxplot(x='Group', y='Score', data=data, palette='pastel')
plt.title('One-Way ANOVA: Group Mean Comparison')
plt.xlabel('Group')
plt.ylabel('Score')
plt.grid(True)
plt.show()

```

```python
#Perform a hypothesis test for population variance using a Chi-square distribution and interpret the result
import numpy as np
from scipy.stats import chi2

np.random.seed(0)

sample = np.random.normal(loc=50, scale=5, size=30)
sample_size = len(sample)
sample_variance = np.var(sample, ddof=1)

hypothesized_variance = 25

chi2_stat = (sample_size - 1) * sample_variance / hypothesized_variance
df = sample_size - 1
p_value = 2 * min(chi2.cdf(chi2_stat, df), 1 - chi2.cdf(chi2_stat, df))

print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (variance is significantly different).")
else:
    print("Result: Fail to reject the null hypothesis (variance is not significantly different).")
```

```python
#Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

success_a = 40
nobs_a = 100

success_b = 55
nobs_b = 120

count = np.array([success_a, success_b])
nobs = np.array([nobs_a, nobs_b])

z_stat, p_value = proportions_ztest(count, nobs)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (proportions are significantly different).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference in proportions).")
```

```python
#D Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

np.random.seed(0)

data1 = np.random.normal(50, 5, 30)
data2 = np.random.normal(50, 10, 30)

var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

f_stat = var1 / var2 if var1 > var2 else var2 / var1
df1 = len(data1) - 1
df2 = len(data2) - 1

p_value = 2 * (1 - f.cdf(f_stat, df1, df2)) if var1 > var2 else 2 * f.cdf(f_stat, df1, df2)

print(f"Variance 1: {var1:.4f}")
print(f"Variance 2: {var2:.4f}")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (variances are significantly different).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference in variances).")

plt.figure(figsize=(8, 5))
plt.hist(data1, bins=10, alpha=0.6, label='Dataset 1', color='skyblue')
plt.hist(data2, bins=10, alpha=0.6, label='Dataset 2', color='salmon')
plt.title('Distribution of Two Datasets')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Perform a Chi-square test for goodness of fit with simulated data and analyze the results
import numpy as np
from scipy.stats import chisquare

np.random.seed(0)

observed = np.random.choice([1, 2, 3, 4], size=100, p=[0.1, 0.2, 0.3, 0.4])
observed_counts = np.bincount(observed)[1:]  # ignore index 0

expected_probs = np.array([0.1, 0.2, 0.3, 0.4])
expected_counts = expected_probs * np.sum(observed_counts)

chi2_stat, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

print(f"Observed Counts: {observed_counts}")
print(f"Expected Counts: {np.round(expected_counts, 2)}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (observed does not fit expected distribution).")
else:
    print("Result: Fail to reject the null hypothesis (observed fits expected distribution).")
```
