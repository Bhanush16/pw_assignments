#SVM & Naive bayes Assignment


### **1. What is a Vector Machine (SVM)?**

SVM is a supervised machine-learning algorithm used for classification and regression.
It finds the **best separating hyperplane** that maximizes the **margin** between classes.

---

### **2. Difference between Hard Margin and Soft Margin SVM**

* **Hard Margin SVM**

  * Assumes data is **perfectly linearly separable**.
  * No misclassification allowed.
  * Very sensitive to noise/outliers.

* **Soft Margin SVM**

  * Allows some misclassification using slack variables.
  * Works even with non-separable data.
  * Controlled by parameter **C**.

---

### **3. Mathematical intuition behind SVM**

SVM tries to:

* Maximize margin ⇒ maximize distance between hyperplane and closest points.
* Subject to class separation constraints.
  Formulated as a **convex optimization** problem that finds the optimal hyperplane.

---

### **4. Role of Lagrange Multipliers in SVM**

Lagrange multipliers convert constrained optimization into an unconstrained dual problem.
They help:

* Identify support vectors
* Enable efficient optimization
* Allow kernel trick usage.

---

### **5. What are Support Vectors in SVM?**

Support vectors are the **data points closest to the hyperplane**.
They determine:

* Margin width
* Position of the decision boundary
  Removing them changes the classifier.

---

### **6. What is a Support Vector Classifier (SVC)?**

SVC is the classification implementation of SVM.
It finds the optimal hyperplane that separates classes with maximum margin.

---

### **7. What is a Support Vector Regressor (SVR)?**

SVR is the regression version of SVM.
It fits a function within an **epsilon-tube** and tries to keep errors within tolerance.

---

### **8. What is the Kernel Trick in SVM?**

Kernel trick maps data into a higher-dimensional feature space **without computing it explicitly**.
This allows SVM to learn **non-linear decision boundaries**.

---

### **9. Compare Linear, Polynomial, and RBF Kernel**

| Kernel         | When to Use                  | Behavior                                  |
| -------------- | ---------------------------- | ----------------------------------------- |
| **Linear**     | Data is linearly separable   | Straight hyperplane                       |
| **Polynomial** | Interaction between features | Curved boundaries, controlled by degree   |
| **RBF**        | Most common, non-linear data | Highly flexible, smooth decision boundary |

---

### **10. Effect of the C Parameter in SVM**

* **High C**: Low bias, high variance → tries to classify all points correctly → risky overfitting
* **Low C**: High bias, low variance → allows misclassification → smoother boundary

---

### **11. Role of Gamma Parameter in RBF Kernel**

Gamma controls how far the influence of a training point spreads:

* **High gamma**: Very narrow influence → overfitting
* **Low gamma**: Wide influence → underfitting

---

## **Naïve Bayes**

### **12. What is the Naïve Bayes classifier, and why is it called “Naïve”?**

Naïve Bayes is a probabilistic classifier based on **Bayes’ theorem**.
It is called “naïve” because it assumes **all features are independent**, which is rarely true.

---

### **13. What is Bayes’ Theorem?**

[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
]
It calculates the probability of event A given B.

---

### **14. Differences between Gaussian, Multinomial, and Bernoulli NB**

| Variant            | Used For            | Feature Type                                     |
| ------------------ | ------------------- | ------------------------------------------------ |
| **Gaussian NB**    | Continuous data     | Real-valued features (e.g., height, temperature) |
| **Multinomial NB** | Text classification | Word counts / frequencies                        |
| **Bernoulli NB**   | Binary features     | Word present/absent                              |

---

### **15. When to use Gaussian Naïve Bayes?**

Use Gaussian NB when features are **continuous** and follow a **normal distribution** (or approximately so).

---

### **16. Key assumptions made by Naïve Bayes**

* Features are **conditionally independent** given the class.
* Each feature contributes **equally and independently**.

---

### **17. Advantages and Disadvantages of Naïve Bayes**

**Advantages**

* Very fast
* Works well with high-dimensional data
* Performs well with text
* Requires small training data

**Disadvantages**

* Strong independence assumption (often unrealistic)
* Poor performance when features are correlated

---

### **18. Why is Naïve Bayes good for text classification?**

* Text data is high-dimensional → NB handles it well
* Word independence assumption works surprisingly well
* Simple and fast
* Works well with sparse data (bag-of-words)

---

### **19. Compare SVM and Naïve Bayes for classification**

| Aspect                  | SVM                                   | Naïve Bayes      |
| ----------------------- | ------------------------------------- | ---------------- |
| **Type**                | Discriminative                        | Generative       |
| **Speed**               | Slower                                | Very fast        |
| **Data**                | Works with high-dimension, non-linear | Best for text    |
| **Interpretability**    | Medium                                | High             |
| **Accuracy**            | Often higher                          | Good but may lag |
| **Feature correlation** | Handles well                          | Struggles        |

---

### **20. How does Laplace Smoothing help in Naïve Bayes?**

Laplace smoothing adds **1** to word counts to avoid zero probabilities.
It prevents the entire probability from becoming zero when a word **never appeared** in the training data.

---

```python
# Train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary

import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
import numpy as np

X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=42)
clf = SVC(kernel='poly', degree=3, C=1).fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, Z, colors='k', levels=[0], linewidths=2)
plt.show()
```

```python
# Train a Multinomial Naïve Bayes classifier for text classification using the 20 Newsgroups dataset

from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

data = fetch_20newsgroups(subset='all')
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)

vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

model = MultinomialNB()
model.fit(X_train_vec, y_train)

pred = model.predict(X_test_vec)
print("Accuracy:", accuracy_score(y_test, pred))
```

```python
# Train an SVM Classifier with different C values and compare the decision boundaries visually

import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
import numpy as np

X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=42)

C_values = [0.1, 1, 10]
plt.figure(figsize=(12, 4))

for i, C in enumerate(C_values):
    clf = SVC(kernel='linear', C=C).fit(X, y)

    plt.subplot(1, 3, i+1)
    plt.title(f"C = {C}")
    plt.scatter(X[:, 0], X[:, 1], c=y)

    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = clf.decision_function(xy).reshape(XX.shape)

    ax.contour(XX, YY, Z, colors='k', levels=[0], linewidths=2)

plt.tight_layout()
plt.show()
```

```python
# Apply feature scaling before training an SVM model and compare results with unscaled data

from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_unscaled = SVC()
svm_unscaled.fit(X_train, y_train)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm_scaled = SVC()
svm_scaled.fit(X_train_scaled, y_train)

print("Unscaled Accuracy:", accuracy_score(y_test, svm_unscaled.predict(X_test)))
print("Scaled Accuracy:", accuracy_score(y_test, svm_scaled.predict(X_test_scaled)))
```

```python
# Train a Gaussian Naïve Bayes model and compare the predictions before and after Laplace Smoothing

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model1 = GaussianNB()
model1.fit(X_train, y_train)

model2 = GaussianNB(var_smoothing=1e-2)
model2.fit(X_train, y_train)

print("Default Smoothing Accuracy:", accuracy_score(y_test, model1.predict(X_test)))
print("Higher Smoothing Accuracy:", accuracy_score(y_test, model2.predict(X_test)))
```

```python
# Train an SVM Classifier on an imbalanced dataset, apply class weighting, and check if it improves accuracy

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
from collections import Counter

# create imbalanced data
X, y = make_classification(n_samples=2000, n_features=20, n_classes=2,
                           weights=[0.95, 0.05], flip_y=0, random_state=42)
print("Class distribution:", Counter(y))

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=42)

# without class weights
clf_plain = SVC(kernel='rbf', C=1, probability=True, random_state=42)
clf_plain.fit(X_train, y_train)
pred_plain = clf_plain.predict(X_test)

# with class weights
clf_weighted = SVC(kernel='rbf', C=1, class_weight='balanced', probability=True, random_state=42)
clf_weighted.fit(X_train, y_train)
pred_weighted = clf_weighted.predict(X_test)

print("Plain Accuracy:", accuracy_score(y_test, pred_plain))
print("Weighted Accuracy:", accuracy_score(y_test, pred_weighted))
print("Weighted Classification Report:\n", classification_report(y_test, pred_weighted))
```

```python
# Implement a Naïve Bayes classifier for spam detection using email data stored in folders 'spam' and 'ham'
# Assumes a local folder 'spam_data' with subfolders 'spam' and 'ham' containing text files

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

texts = [
    "Free money, click here",          # spam
    "Win a lottery now",               # spam
    "Hi, how are you doing today?",    # ham
    "Let's meet for coffee tomorrow",  # ham
    "Congratulations — you won!"       # spam
]
labels = [1, 1, 0, 0, 1]  # 1 = spam, 0 = ham

X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.4, random_state=42)

vec = CountVectorizer()
X_tr = vec.fit_transform(X_train)
X_te = vec.transform(X_test)

model = MultinomialNB()
model.fit(X_tr, y_train)
pred = model.predict(X_te)

print("Accuracy:", accuracy_score(y_test, pred))

```

```python
# Train an SVM Classifier and a Naïve Bayes Classifier on the same dataset and compare their accuracy

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

data = load_wine()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

svm = SVC(kernel='rbf', probability=True, random_state=42)
nb = GaussianNB()

svm.fit(X_train, y_train)
nb.fit(X_train, y_train)

print("SVM Accuracy:", accuracy_score(y_test, svm.predict(X_test)))
print("GaussianNB Accuracy:", accuracy_score(y_test, nb.predict(X_test)))
```

```python
# Perform feature selection before training a Naïve Bayes classifier and compare results (SelectKBest + chi2)

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

data = load_digits()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# without feature selection
nb_plain = MultinomialNB()
nb_plain.fit(X_train, y_train)

# with feature selection
selector = SelectKBest(chi2, k=30)
X_train_sel = selector.fit_transform(X_train, y_train)
X_test_sel = selector.transform(X_test)

nb_sel = MultinomialNB()
nb_sel.fit(X_train_sel, y_train)

print("No selection Accuracy:", accuracy_score(y_test, nb_plain.predict(X_test)))
print("With SelectKBest Accuracy:", accuracy_score(y_test, nb_sel.predict(X_test_sel)))
```

```python
# Train SVM using One-vs-Rest (OvR) and One-vs-One (OvO) on the Wine dataset and compare accuracy

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

data = load_wine()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

ovr = OneVsRestClassifier(SVC(kernel='linear', probability=True))
ovo = OneVsOneClassifier(SVC(kernel='linear', probability=True))

ovr.fit(X_train, y_train)
ovo.fit(X_train, y_train)

print("OvR Accuracy:", accuracy_score(y_test, ovr.predict(X_test)))
print("OvO Accuracy:", accuracy_score(y_test, ovo.predict(X_test)))
```

```python
# Train SVM with Linear, Polynomial, and RBF kernels on the Breast Cancer dataset and compare accuracy

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

data = load_breast_cancer()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

kernels = {'linear': SVC(kernel='linear', probability=True),
           'poly': SVC(kernel='poly', degree=3, probability=True),
           'rbf': SVC(kernel='rbf', probability=True)}

for name, clf in kernels.items():
    clf.fit(X_train, y_train)
    print(f"{name} Accuracy:", accuracy_score(y_test, clf.predict(X_test)))
```

```python
# Train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the average accuracy

from sklearn.datasets import load_iris
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.svm import SVC
import numpy as np

data = load_iris()
X, y = data.data, data.target
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
clf = SVC(kernel='rbf')

scores = cross_val_score(clf, X, y, cv=skf, scoring='accuracy')
print("Fold accuracies:", scores)
print("Average accuracy:", np.mean(scores))
```

```python
# Train a Naïve Bayes classifier using different prior probabilities and compare performance (GaussianNB)

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

data = load_breast_cancer()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# default priors (estimated)
gnb_default = GaussianNB()
gnb_default.fit(X_train, y_train)

# custom priors (e.g., bias towards class 0)
gnb_custom = GaussianNB(priors=[0.7, 0.3])
gnb_custom.fit(X_train, y_train)

print("Default priors Accuracy:", accuracy_score(y_test, gnb_default.predict(X_test)))
print("Custom priors Accuracy:", accuracy_score(y_test, gnb_custom.predict(X_test)))
```

```python
# Perform Recursive Feature Elimination (RFE) before training an SVM Classifier and compare accuracy

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score

data = load_wine()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

svc = SVC(kernel='linear')
svc.fit(X_train, y_train)
plain_acc = accuracy_score(y_test, svc.predict(X_test))

rfe = RFE(estimator=SVC(kernel='linear'), n_features_to_select=6, step=1)
rfe.fit(X_train, y_train)
svc_rfe = SVC(kernel='linear')
svc_rfe.fit(rfe.transform(X_train), y_train)
rfe_acc = accuracy_score(y_test, svc_rfe.predict(rfe.transform(X_test)))

print("Plain SVM Accuracy:", plain_acc)
print("RFE + SVM Accuracy:", rfe_acc)
```

```python
# Train an SVM Regressor (SVR) and evaluate its performance using Mean Absolute Error (MAE) instead of MSE

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error

data = fetch_california_housing()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svr = SVR(kernel='rbf')
svr.fit(X_train, y_train)
pred = svr.predict(X_test)

print("MAE:", mean_absolute_error(y_test, pred))
```

```python
# Train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC score (binary)

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score

data = load_breast_cancer()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

gnb = GaussianNB()
gnb.fit(X_train, y_train)
probs = gnb.predict_proba(X_test)[:, 1]

print("ROC-AUC:", roc_auc_score(y_test, probs))
```

```python
# Train an SVM Classifier and visualize the Precision-Recall Curve (binary)

import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, average_precision_score

X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, weights=[0.8,0.2], random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

clf = SVC(kernel='rbf', probability=True)
clf.fit(X_train, y_train)
probs = clf.predict_proba(X_test)[:, 1]

precision, recall, _ = precision_recall_curve(y_test, probs)
ap = average_precision_score(y_test, probs)

plt.plot(recall, precision, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title(f'Precision-Recall curve (AP={ap:.3f})')
plt.show()
```
